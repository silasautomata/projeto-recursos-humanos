```python
# Libs Necessárias

# Libs para Modelagem e Matrizez
import numpy as np
import pandas as pd

# Libs para anaálises gráficas
import matplotlib.pyplot as plt
import seaborn as sns

# Lib para ignorar avisos
import warnings

# Desabilitando avisos
warnings.filterwarnings('ignore')
```

```python
base_dados = pd.read_csv('/content/Salary_Data.csv')

base_dados.head()
```

```python
# renomeando as colunas
base_dados.rename(columns={
    'YearsExperience': 'xp',
    'Salary':'renda'
}, inplace=True)
```

```python
base_dados.shape # linhas e colunas
```

```python
# campos nulos
base_dados.isnull().sum()
```

```python
# grafico nulos
sns. heatmap(base_dados.isnull(), cbar=False);
```

```python
base_dados.describe() # estatistica
```

```python
plt.figure(figsize=(15,6)) # tamanhho da figura
plt.title('analise da renda') # titulo
sns.kdeplot(base_dados['renda'], color='red'); # grafico da coliuna de renda
```

```python
plt.figure(figsize=(10,6)) # tamanho da figura
plt.title('analise da xp') # titulo
sns.kdeplot(base_dados['xp'], color='blue'); # gafico da coluna de experiencioa xp
```

```python
# distribuiçao
plt.figure(figsize=(10, 5)); # tamando da figura
sns.boxplot(base_dados['renda']); # grafico de renda boxplot
```

```python
# distribuiçao
plt.figure(figsize=(10, 5)) # tamanho da figura
sns.boxplot(base_dados['xp']); # grafico boxplot da coluna xp
```

```python
plt.figure(figsize=(10, 5)) # tamanho da figura
sns.scatterplot(data=base_dados, x='renda', y='xp'); # grafico escaterplot eixo X renda e eixo y experiencioa xp
```

```python
plt.figure(figsize=(10,5)) # tamanho da figura
sns.regplot(data=base_dados, x='renda', y='xp'); # grafico regplot para gerar uma linha no grafico
```

```python
# correlaçao
correlacao = base_dados.corr('pearson') # correlaçao de pearson
plt.figure(figsize=(10, 5)) # tamanho das figura
# seaborn- matriz de correlaçao- mostra os valores numericos - exibe a barra de cores ao lado direito
sns.heatmap( correlacao, annot=True, cbar=True);
```

```python
correlacao = base_dados.corr('spearman') # correlaçao de spearman
plt.figure(figsize=(10, 5))
sns.heatmap(correlacao, annot=True, cbar=True);
```
